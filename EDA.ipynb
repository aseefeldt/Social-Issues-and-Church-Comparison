## EDA
making the column names for same for the states, and making the full name to be the same across all data sets.  

```python
churches = pd.read_csv('./data/2020_church.csv')
```

```python
churches.rename(columns={'State Name': 'state'}, inplace=True)
```

```python
churches.head()
```

```python
homeless = pd.read_csv('./data/2022_homeless.csv')
```

```python
def convert_state_code(state_code):
    state_codes = {
        'AL': 'Alabama',
        'AK': 'Alaska',
        'AZ': 'Arizona',
        'AR': 'Arkansas',
        'CA': 'California',
        'CO': 'Colorado',
        'CT': 'Connecticut',
        'DE': 'Delaware',
        'FL': 'Florida',
        'GA': 'Georgia',
        'HI': 'Hawaii',
        'ID': 'Idaho',
        'IL': 'Illinois',
        'IN': 'Indiana',
        'IA': 'Iowa',
        'KS': 'Kansas',
        'KY': 'Kentucky',
        'LA': 'Louisiana',
        'ME': 'Maine',
        'MD': 'Maryland',
        'MA': 'Massachusetts',
        'MI': 'Michigan',
        'MN': 'Minnesota',
        'MS': 'Mississippi',
        'MO': 'Missouri',
        'MT': 'Montana',
        'NE': 'Nebraska',
        'NV': 'Nevada',
        'NH': 'New Hampshire',
        'NJ': 'New Jersey',
        'NM': 'New Mexico',
        'NY': 'New York',
        'NC': 'North Carolina',
        'ND': 'North Dakota',
        'OH': 'Ohio',
        'OK': 'Oklahoma',
        'OR': 'Oregon',
        'PA': 'Pennsylvania',
        'RI': 'Rhode Island',
        'SC': 'South Carolina',
        'SD': 'South Dakota',
        'TN': 'Tennessee',
        'TX': 'Texas',
        'UT': 'Utah',
        'VT': 'Vermont',
        'VA': 'Virginia',
        'WA': 'Washington',
        'WV': 'West Virginia',
        'WI': 'Wisconsin',
        'WY': 'Wyoming'
    }
    
    return state_codes.get(state_code.upper(), 'Invalid State Code')
homeless['state'] = homeless['state'].apply(convert_state_code)
```

```python
homeless.head()
```

```python
foster = pd.read_csv('./data/foster_care.csv')
```

```python
foster.rename(columns={'Puerto Rico': 'state', '2259': 'foster'}, inplace=True)
```

```python
foster.head()
```

```python
uninsured = pd.read_csv('./data/uninsured-2021.csv')
uninsured.rename(columns={'State Name': 'state'}, inplace=True)
```

```python
uninsured.head()
```

```python
food = pd.read_csv('./data/food_insecure.csv')
food['state'] = food['state'].apply(convert_state_code)
```

```python
food.head()
```

```python
housing = pd.read_csv('./data/housing_burdened.csv')
```

```python
housing.head()
```

## Merging all data sets into one.  

```python
df = pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(churches, homeless,on='state'), foster, on='state'), uninsured, on='state'), food, on='state'), housing, on='state')
df.head()
```

```python
#exporting the data set

df.to_csv('./data/merged.csv')
```
